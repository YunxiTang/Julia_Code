Constrained Optimization

```julia
using ForwardDiff
using PyPlot
using LinearAlgebra
```

```julia
Q = Diagonal([0.5;1.0])
function f(x)
    return 0.5*(x-[1;0])'*Q*(x-[1;0])
end
function ∇f(x)
    return Q*(x-[1;0])
end
function ∇2f(x)
    return Q
end
```

```julia
function c(x)
    # constraint
    return x[1]^2 + 2*x[1] - x[2]
end

function ∂c(x)
    return [2*x[1]+2 -1]
end
```

```julia
function plot_landscape()
    Nsamp = 50
    Xsamp = kron(ones(Nsamp),LinRange(-4,4,Nsamp)')
    Ysamp = kron(ones(Nsamp)',LinRange(-4,4,Nsamp))
    Zsamp = zeros(Nsamp, Nsamp)
    for j=1:Nsamp
        for k=1:Nsamp
            Zsamp[j,k] = f([Xsamp[j,k]; Ysamp[j,k]])
        end
    end
    contour(Xsamp, Ysamp, Zsamp)
    
    xc = LinRange(-3.2,1.2,Nsamp)
    plot(xc,xc.^2+2.0 .* xc, "y")
end
plot_landscape()
```

```julia
function newton_step(x0,λ0)
    H = ∇2f(x0) + ForwardDiff.jacobian(x -> ∂c(x)'*λ0, x0)
    β = 0.3
    # add regularization (IMPORTANT!!!)
    while !isposdef(H)
        H = H + β .* I
    end
    C = ∂c(x0)
    Δz = [H C'; C 0]\[-∇f(x0)-C'*λ0; -c(x0)]
    Δx = Δz[1:2]
    Δλ = Δz[3]
    return x0+Δx, λ0.+Δλ
end
```

```julia
x_init = [-3;1]
λ_init = [0.0]
plot_landscape()
plot(x_init[1],x_init[2],"bs")
x_new, λ_new = newton_step(x_init, λ_init)
x_guess = [x_init x_new]
λ_guess = [λ_init λ_new]
for i=1:10
    x_new, λ_new = newton_step(x_new, λ_new)
    x_guess = [x_guess x_new]
    λ_guess = [λ_init λ_new]
end
plot(x_guess[1,:], x_guess[2,:],"kx")
plot(x_guess[1,end], x_guess[1,end],"rx")
display(x_new)
```

```julia
function gauss_newton_step(x0,λ0)
    H = ∇2f(x0)
    C = ∂c(x0)
    Δz = [H C'; C 0]\[-∇f(x0)-C'*λ0; -c(x0)]
    Δx = Δz[1:2]
    Δλ = Δz[3]
    return x0+Δx, λ0.+Δλ
end
```

```julia
xguess = [-3; 2.0]
λguess = [0.0]
plot_landscape()
plot(xguess[1], xguess[2], "rx")
```

```julia
xnew, λnew = gauss_newton_step(xguess[:,end],λguess[end])
xguess = [xguess xnew]
λguess = [λguess λnew]
plot_landscape()
plot(xguess[1,:], xguess[2,:], "rx")
```

```julia
x_init = [-3;1]
λ_init = [0.0]
plot_landscape()
plot(x_init[1],x_init[2],"bs")
x_new, λ_new = gauss_newton_step(x_init, λ_init)
x_guess = [x_init x_new]
for i=1:200
    x_new, λ_new = gauss_newton_step(x_new, λ_new )
    x_guess = [x_guess x_new]
end
plot(x_guess[1,:], x_guess[2,:],"gx")
plot(x_guess[1,end], x_guess[1,end],"rx")
display(x_guess[:,end])
```
