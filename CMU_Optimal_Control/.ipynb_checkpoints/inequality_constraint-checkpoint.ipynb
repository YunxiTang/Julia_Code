Inequality constraint

```julia
using LinearAlgebra
using PyPlot
using ForwardDiff
```

```julia
Q = Diagonal([0.5; 1.0])
function f(x)
    # cost function
    return 0.5*(x-[1; 0])'*Q*(x-[1; 0])
end

function ∇f(x)
    return Q*(x-[1; 0])
end

function ∇2f(x)
    return Q
end
```

```julia
A = [1.0 -1.0]
b = -1.0
function c(x)
    # inequality constraint
    return dot(A,x)-b
end

function ∂c(x)
    return A
end
```

```julia
function plot_landscape()
    Nsamp = 50
    Xsamp = kron(ones(Nsamp),LinRange(-4,4,Nsamp)')
    Ysamp = kron(ones(Nsamp)',LinRange(-4,4,Nsamp))
    Zsamp = zeros(Nsamp,Nsamp)
    for j = 1:Nsamp
        for k = 1:Nsamp
            Zsamp[j,k] = f([Xsamp[j,k]; Ysamp[j,k]])
        end
    end
    contour(Xsamp,Ysamp,Zsamp)

    xc = LinRange(-4,3,Nsamp)
    plot(xc,xc.+1,"k")
end

plot_landscape()
```

```julia
function La(x, λ, ρ)
    # Augmented Lagrangian function
    p = max(0, c(x))
    return f(x) + λ*p + (ρ/2)*(p'*p)
end
```

```julia
function newton_solve(x0,λ,ρ)
    x = x0
    p = max.(0, c(x))
    C = zeros(1,2)
    if 0 ≤ c(x)
        C = ∂c(x)
    end
    g = ∇f(x) + (λ+ρ*p)*C'
    while norm(g) ≥ 1e-8
        H = ∇2f(x) + ρ*C'*C
        Δx = -H\g
        
        x = x+Δx
        
        p = max.(0,c(x))
        C = zeros(1,2)
        if c(x) ≥ 0
            C = ∂c(x)
        end
        g = ∇f(x) + (λ+ρ*p)*C'
    end
    
    return x
end
```
