Inequality constraint

```julia
using LinearAlgebra
using PyPlot
using ForwardDiff
```

```julia
Q = Diagonal([0.5; 1.0])
function f(x)
    # cost function
    return 0.5*(x-[1; 0])'*Q*(x-[1; 0])
end

function ∇f(x)
    return Q*(x-[1; 0])
end

function ∇2f(x)
    return Q
end
```

```julia
A = [1.0 -1.0; 1.0 1.0]
b = [-1.0; 0.0]
function c(x)
    # inequality constraint
    return A*x-b
end

function ∂c(x)
    return A
end
```

```julia
function plot_landscape()
    Nsamp = 50
    Xsamp = kron(ones(Nsamp),LinRange(-4,4,Nsamp)')
    Ysamp = kron(ones(Nsamp)',LinRange(-4,4,Nsamp))
    Zsamp = zeros(Nsamp,Nsamp)
    for j = 1:Nsamp
        for k = 1:Nsamp
            Zsamp[j,k] = f([Xsamp[j,k]; Ysamp[j,k]])
        end
    end
    contour(Xsamp,Ysamp,Zsamp)

    xc = LinRange(-4,3,Nsamp)
    plot(xc, -xc.+ 1.0,"r")
    plot(xc, xc .+ 1,"k")
end

plot_landscape()
```

```julia
function La(x, λ, ρ)
    # Augmented Lagrangian function
    p = max.(0, c(x))
    return f(x) + λ'*p + (ρ/2)*(p'*p)
end
```

```julia
function newton_solve(x0,λ,ρ)
    x = x0
    p = max.(0, c(x))
    C = zeros(2,2)
    if 0 ≤ c(x)[1] 
        C[1,:] = ∂c(x)[1,:]
    elseif 0 ≤ c(x)[2]
        C[2,:] = ∂c(x)[2,:]
    end
    g = ∇f(x) + C*(λ.+ρ*p)
    while norm(g) ≥ 1e-8
        H = ∇2f(x) + ρ*C'*C
        Δx = -H\g
        
        x = x+Δx
        
        p = max.(0,c(x))
        C = zeros(2,2)
        if 0 ≤ c(x)[1] 
            C[1,:] = ∂c(x)[1,:]
        elseif 0 ≤ c(x)[2]
            C[2,:] = ∂c(x)[2,:]
        end
        g = ∇f(x) + C*(λ.+ρ*p)
    end
    
    return x
end
```

```julia
xguess = [-3; 2]
λguess = [0.0;0.0]
ρ = 1.0
plot_landscape()
plot(xguess[1], xguess[2], "rx")
```

```julia
xnew = newton_solve(xguess[:,end],λguess[:,end],ρ)
λnew = max.(0,λguess[:,end]+ρ*c(xnew))
xguess = [xguess xnew]
λguess = [λguess λnew]
for i=1:5
    ρ = 10*ρ
    xnew = newton_solve(xguess[:,end],λguess[:,end],ρ)
    xguess = [xguess xnew]
    λguess = [λguess λnew]
end
plot_landscape()
plot(xguess[1,:], xguess[2,:], "rx")
```
